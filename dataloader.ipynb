```python
import cv2
import numpy as np
import pandas as pd
import os
def pad_image(image):
    img_path = os.path.expanduser(image)
    img = cv2.imread(img_path)
    if img is None:
        print("imread error. PATH: ",img_path)
        return None
    print(img.shape)
    
    h, w, c = img.shape
    top = bottom = left = right = 0
    h1 =  h % 640
    w1 = w % 640
    
    if h1 > 0 or w1> 0:
        bottom = 640 - h1
        right = 640 - w1
        img = cv2.copyMakeBorder(
        img, top, bottom, left, right,
        borderType=cv2.BORDER_CONSTANT,
        value=[0, 0, 0]  # black padding
        )
    print(img.shape)
    
    cv2.imshow("image", img)
    cv2.waitKey(0)        # Waits for a key press
    cv2.destroyAllWindows()  # Closes the window

    return 
```

```python
def split_img(image):
    img_path = os.path.expanduser(image)
    img = cv2.imread(img_path)
    if img is None:
        print("imread error. PATH: ",img_path)
        return None
    print(img.shape)
    h, w = img.shape[:2]
    tile_size = 640
    tiles = []
    for y in range(0, h, tile_size):
        for x in range(0, w, tile_size):
            tile = img[y:y+tile_size, x:x+tile_size]
            tiles.append(tile)
    return tiles
        
    
```
